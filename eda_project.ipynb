```python
df = pd.read_csv('C:\\Users\\Admin\\Downloads\\archive (8)\\pizza_sales.csv')
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True, errors='coerce')
print(df[df['order_date'].isna()])

```

```python
# 1 )  which pizza is sold the most ?
sales_summary=df.groupby('pizza_name')['quantity'].sum().reset_index()
most_sold=sales_summary.loc[sales_summary['quantity'].idxmax()]
print(f"the pizza that is sold the most {most_sold}with{sales_summary}sales.")
plt.figure(figsize=(10,6))
plt.bar(sales_summary['pizza_name'],sales_summary['quantity'],color='blue')
plt.axhline(y=most_sold['quantity'],color='r',linestyle='--',label='Most Sold Pizza')
plt.title('Pizza Sales Summary')
plt.xlabel('pizza_name')
plt.ylabel('quantity')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show
```

```python
#conclusion:the most sold pizza is the classic deluxe pizza followed by the barbecue chicken pizza and the hawaiian pizza 
```

```python
# 2)  which ingredients is used the most?
from collections import Counter
all_ingredients = []
for ingredients in df['pizza_ingredients']:
    ingredient_list = [ingredient.strip().lower() for ingredient in ingredients.split(",")]
    all_ingredients.extend(ingredient_list)
ingredient_counts = Counter(all_ingredients)
ingredients = list(ingredient_counts.keys())
counts = list(ingredient_counts.values())
print(f"the ingredients that is used the most {ingredients}with{counts}sales.")


plt.figure(figsize=(10, 6))
plt.bar(ingredients, counts, color='skyblue')
plt.xlabel('Ingredients')
plt.ylabel('Number of Occurrences')
plt.title('Ingredient Usage in Pizza Sales')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Conclusion: Cheese, Pepperoni, and Chicken are likely the most-used ingredients due to their presence in high-selling pizza categories.
```

```python
# 3)  which pizza_category is sold the most ?
category_counts = df['pizza_category'].value_counts()
categories = category_counts.index
counts = category_counts.values
print(f"the most sold pizza category {categories}with{counts}sales.")
plt.figure(figsize=(10, 6))
plt.pie(counts,labels=category_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Pizza sold time')
plt.axis('equal')
plt.show()
```

```python
#Conclusion: While chicken-based pizzas perform strongly, classic and traditional pizzas slightly outperform chicken varieties, led by The Classic Deluxe, Hawaiian, and Pepperoni Pizzas.
```

```python
# 4) during which has the pizza sold the most ? morning ,afternoon,evening and night ?
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('C:\\Users\\Admin\\Downloads\\archive (8)\\pizza_sales.csv')
def categorize_time(order_time):
    hour =pd.to_datetime(order_time).hour
    if 6 <= hour < 12:
        return'morning'
    elif 12 <= hour < 17 :
        return'afternoon'
    elif 17 <= hour < 21 :
        return'evening'
    else:
        return'night'
df['time_of_day']=df['order_time'].apply(categorize_time)
sales_summary=df.groupby('time_of_day').sum()
peak_time=sales_summary.idxmax()
peak_sales=sales_summary.max()
time_counts=df['time_of_day'].value_counts()
print(f"the time of day with the most pizza sales is {peak_time}with{peak_sales}sales.")
plt.figure(figsize=(10, 6))
plt.pie(time_counts,labels=time_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Pizza sold time')
plt.axis('equal')
plt.show()
```

```python
#Conclusion: Most pizza sales happen in the afternoon, but nights are significant for large-sized pizzas and specialty orders.
```

```python
# 6)  what is the average sale revenue of a single day ?

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('C:\\Users\\Admin\\Downloads\\archive (8)\\pizza_sales.csv')
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
daily_revenue = df.groupby('order_date')['total_price'].sum()
overall_average = daily_revenue.mean()
daily_average = daily_revenue.mean()
print('Daily average revenue:', daily_average)
print('Overall average revenue per day:', overall_average)

# Plotting the daily revenue as a histogram
plt.figure(figsize=(12, 6))
plt.hist(daily_revenue, bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Daily Pizza Sales Revenue', fontsize=16)
plt.xlabel('Total Revenue ($)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.tight_layout() 
plt.show()

```

```python
# 7)  on which day has the mist sale occured ?
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('C:\\Users\\Admin\\Downloads\\archive (8)\\pizza_sales.csv')

# Convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')

# Group by order_date and sum the total_price
daily_sales = df.groupby('order_date')['total_price'].sum().reset_index()

# Identify the day with the maximum sales
max_sales_index = daily_sales['total_price'].idxmax()
max_sales_date = daily_sales.loc[max_sales_index, 'order_date']
max_sales_value = daily_sales.loc[max_sales_index, 'total_price']
print(f'The day with the most sales is {max_sales_date.date()} with total sales of ${max_sales_value:.2f}.')
plt.figure(figsize=(12, 6))
plt.hist(daily_sales['total_price'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Daily Pizza Sales Revenue', fontsize=16)
plt.xlabel('Total Revenue ($)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.tight_layout()  # Adjust layout to make room for the labels
plt.show()
```

```python
# 9)  what is average sales revenue of a week ? from 01-01-2015 to 07-01-2015 

import pandas as pd
import matplotlib.pyplot as plt
df['order_date'] = pd.to_datetime(df['order_date'])
start_date = '2015-01-01'
end_date = '2015-01-07'
filtered_df = df[(df['order_date'] >= start_date) & (df['order_date'] <= end_date)]
daily_revenue = filtered_df.groupby('order_date')['total_price'].sum().reset_index()
total_revenue = filtered_df['total_price'].sum()
num_days = (pd.to_datetime(end_date) - pd.to_datetime(start_date)).days + 1
average_daily_revenue = total_revenue / num_days
print(f"Total revenue from {start_date} to {end_date}: ${total_revenue:.2f}")
print(f"Average daily revenue: ${average_daily_revenue:.2f}")

plt.figure(figsize=(10, 6))
plt.bar(daily_revenue['order_date'], daily_revenue['total_price'], color='skyblue')
plt.title('Daily Sales Revenue from 01-01-2015 to 07-01-2015')
plt.xlabel('Date')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=90)  
plt.tight_layout()  
plt.grid(axis='y')  
plt.show()
```

```python
#Conclusion:The average weekly sales revenue from January 1, 2015, to January 7, 2015, is $17,549.65, with a daily average of $2,507.09. This indicates a steady revenue stream during the week, reflecting consistent customer demand and reliable sales performance over a short-term period.
```

```python
# 10 )  What is the standard deviation of the unit prices of pizzas? Does this indicate a wide range of prices among the pizzas?

import pandas as pd
df['unit_price'] = pd.to_numeric(df['unit_price'], errors='coerce')
std_dev_unit_price = df['unit_price'].std()
mean_unit_price = df['unit_price'].mean()
print(f"Standard Deviation of Unit Prices: ${std_dev_unit_price:.2f}")
print(f"Mean Unit Price: ${mean_unit_price:.2f}")
if std_dev_unit_price > mean_unit_price:
    interpretation = "This indicates a wide range of prices among the pizzas."
else:
    interpretation = "This indicates a narrow range of prices among the pizzas."
print(interpretation)

#plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['unit_price'].dropna(), bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(mean_unit_price, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: ${mean_unit_price:.2f}')
plt.axvline(mean_unit_price + std_dev_unit_price, color='green', linestyle='dashed', linewidth=1.5, label=f'Mean + 1 Std Dev: ${mean_unit_price + std_dev_unit_price:.2f}')
plt.axvline(mean_unit_price - std_dev_unit_price, color='orange', linestyle='dashed', linewidth=1.5, label=f'Mean - 1 Std Dev: ${mean_unit_price - std_dev_unit_price:.2f}')
plt.title('Distribution of Pizza Unit Prices')
plt.xlabel('Unit Price ($)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()  
plt.show()
```

```python
#Conclusion:The standard deviation of unit prices is $3.62, with a mean unit price of $16.49. This reflects a narrow range of prices among the pizzas, indicating a consistent and stable pricing strategy that ensures affordability while appealing to a broad customer base
```
